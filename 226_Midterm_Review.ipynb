## Cauchy's Relationship
Traction vector for a plane defined by the normal vector $\vec{v}$ (that is, $v_j$):

$T_i^\vec{v} = \tau_{ji}v_j$

$T_i^\vec{v}$ will have three components, where $i = 1, 2, 3$. Note $T_i^\vec{v}$ doesn't have to be parallel to $\vec{v}$, its more like extracting the relevant components of traction for the surface normal to $\vec{v}$.

## Conservation of Momentum
$\frac{d}{dt} mv_i=f_i $ (external force)

$⇒ \frac{d}{dt} \int_{V(t)} (ρ v_i)dV
 = \color{orange}{ \int_{S} T_i^v dS}
 + \int_{V(t)} χ_i dV$

$⇒ \frac{d}{dt} \int_{V(t)} (ρ v_i)dV
 = \color{orange}{ \int_{S} τ_{ji}v_j dS}
 + \int_{V(t)} χ_i dV$ (Cauchy)

$⇒ \color{red}{  \frac{d}{dt} \int_{V(t)} (ρ v_i)dV  }
 = \color{orange}{ \int_{V(t)} \frac{∂τ_{ji}}{∂x_j} dV  }
 + \int_{V(t)} χ_i dV$ (Gauss)

$⇒ \color{red}{  \int_{V(t)} \left(
   \frac{d}{dt} ρv_i
 + \frac{∂}{∂x_j}ρ v_i v_j
   \right) dV  }
 = \color{orange}{ \int_{V(t)} \frac{∂τ_{ji}}{∂x_j} dV}
 + \int_{V(t)} χ_i dV $ (material derivative?)

$⇒ \int_{V(t)} \left(
   \color{blue}{  \frac{dρ}{dt}  }v_i
 + \color{magenta}{  ρ\frac{dv_i}{dt} }
 + v_i \color{green}{  \frac{∂(ρ v_j)}{∂x_j} }
 + \color{magenta}{  ρv_j \frac{∂v_i}{∂x_j} }
   \right) dV
 = \color{orange}{  \int_{V(t)} \frac{∂τ_{ji}}{∂x_j} dV  }
 + \int_{V(t)} χ_i dV $ (chain rule)

$⇒ \int_{V(t)}
   \color{magenta}{ ρ ( \frac{dv_i}{dt} }
 + \color{magenta}{  v_j \frac{∂v_i}{∂x_j} ) } dV
 = \int_{V(t)} \left(
   \color{orange}{ \frac{∂τ_{ji}}{∂x_j} }
 + χ_i
   \right) dV $ (from conservation of mass  $⇒ \color{blue}{  \frac{dρ}{dt}  }
   + \color{green}{  \frac{∂(ρ v_j)}{∂x_j} = 0}$;
   combine terms)

Differential form:
$ \color{magenta}{ ρ ( \frac{Dv_i}{Dt} }

# Relating $τ$ and $V_{ij}$ [Lecture 4]

## Velocity Gradient

$ \frac{∂v_i}{∂x_j} =
\begin{pmatrix}
\frac{∂v_1}{∂x_1} & \frac{∂v_1}{∂x_2} & \frac{∂v_1}{∂x_3} \\
\frac{∂v_2}{∂x_1} & \frac{∂v_2}{∂x_2} & \frac{∂v_2}{∂x_3} \\
\frac{∂v_3}{∂x_1} & \frac{∂v_3}{∂x_2} & \frac{∂v_3}{∂x_3}
\end{pmatrix}
 $

Re-writing as:

$ \color{magenta}{ \frac{∂v_i}{∂x_j} }
 = \frac{1}{2} \left(
   \color{magenta}{ \frac{∂v_i}{∂x_j} }
 + \frac{∂v_j}{∂x_i}
   \right)
 - \frac{1}{2} \left(
   \frac{∂v_j}{∂x_i}
 - \color{magenta}{ \frac{∂v_i}{∂x_j} }
   \right)
 = \color{red}{V_{ij}} - Ω_{ij}$

 That is, we can define:

$ \color{red}{V_{ij}}
 = \frac{1}{2} \left(
   \color{magenta}{ \frac{∂v_i}{∂x_j} }
 + \frac{∂v_j}{∂x_i}
   \right)
$, and


$ Ω_{ij}
 = \frac{1}{2} \left(
   \frac{∂v_j}{∂x_i}
 - \color{magenta}{ \frac{∂v_i}{∂x_j} }
   \right)
$

$\color{red}{V_{ij}}$: rate of deformation tensor (or strain rate), $Ω_{ij}$: spin tensor

From material testing, $\color{red}{V_{ij}}$ is related to $τ_{ij}$

## Experiments towards a constitutive equation
* Parallel plates (Couette flow)
* Viscometer
 * Two concentric cylinders
 * Cone and plate

Many fluids have stress $ τ = \frac{Force}{Area} ∝ γ = \frac{u}{h} = \frac{(edge) Velocity}{Thickness} $

Experiments can be summarized with a constitutive equation, valid for any fluid with constant viscosity $μ$ and constant density $ρ$ (incompressibility):

$\begin{equation}\tag{constitutive equation for $τ_{ij}$}
 \color{orange}{τ_{ij}
 = -pδ_{ij}
 + 2μ \color{red}{V_{ij}}}
 \end{equation}
$

$p$: pressure = $ -\frac{1}{3} * τ_{ii}$ = $-\frac{1}{3} * (τ_{11} + τ_{22} + τ_{33}) $

$δ_{ij}$: Kronecker delta = 1 if i=j (that is, only grab the i=j terms $p_{11}, p_{22}, p_{33}$)

$μ$: coefficient of viscosity

## Plugging $\tau_{ij}$ into the conservation of momentum equation to get Navier-Stokes

$ \color{magenta}{ ρ\frac{∂v_i}{∂t} }
 = \color{orange}{ \frac{∂τ_{ji}}{∂x_j} }
 + χ_i
$

$⇒ \color{magenta}{ ρ\frac{∂v_i}{∂t} }
 = \color{orange}{ -\frac{∂}{∂x_j}(p δ_{ji}) }
 + \color{orange}{  \frac{∂}{∂x_j}(2 μ }
 \color{red}{V_{ij})}
 + χ_i
$

$⇒ \color{magenta}{ ρ\frac{∂v_i}{∂t} }
 = \color{orange}{ -\frac{∂}{∂x_j}(p δ_{ji}) }
 + \color{orange}{  \frac{∂}{∂x_j} 2μ  }
   \color{red}{  \frac{1}{2} ( \frac{∂v_i}{∂x_j}
  +\frac{∂v_j}{∂x_i})  }
 + χ_i
$

$⇒ \color{magenta}{ ρ\frac{∂v_i}{∂t} }
 = \color{orange}{ -\frac{∂}{∂x_j}(p δ_{ji}) }
 + \color{orange}{  μ  }
   \color{red}{  \frac{∂^2v_i}{∂x_j^2} }
 + \color{orange}{ μ \frac{∂}{∂x_j}  }
   \color{red}{  \frac{∂v_j}{∂x_i}  }
 + χ_i
$

$⇒ \color{magenta}{ ρ\frac{∂v_i}{∂t} }
 = \color{orange}{ -\frac{∂}{∂x_j}(p δ_{ji}) }
 + \color{orange}{  μ  }
   \color{red}{  \frac{∂^2v_i}{∂x_j^2} }
 + \color{orange}{μ} \frac{\color{orange}{∂}}{\color{red}{∂x_i}}
   \frac{\color{red}{∂v_j}}{\color{orange}{∂x_j} }
 + χ_i
$

Velocity divergence $\frac{\color{red}{∂v_j}}{\color{orange}{∂x_j} } = 0$ for a fluid with constant $ρ$ (density)

$⇒\begin{equation}\tag{Navier-Stokes}
 \color{magenta}{ ρ\frac{dv_i}{dt} }
 = \color{orange}{ -\frac{∂p}{∂x_i}
 + μ\color{red}{\frac{∂^2 v_i}{∂x_j^2}}}
 + χ_i
 \end{equation}
$

4 equations in 4 unknowns (3 equations from i=1,2,3. What about the 4th equation? $\frac{∂v_i}{∂x_i} = 0$?)

This Navier-Stokes expression applies for fluids that are:
* Newtonian: viscosity $μ=constant$ (Wikipedia has a more general defintion?)
 * examples: plasma, urine
* Incompressible: density $ρ = constant ⇒ \frac{\color{red}{∂v_j}}{\color{orange}{∂x_j} } = 0$ from conservation of mass (see practice problems)

## $μ(J_2)$
Update the older constitutive equation to make $μ$ a function of $J_2$:

$\color{orange}{τ_{ij}
 = -pδ_{ij}
 + 2\color{crimson}{μ(J_2)} \color{red}{V_{ij}}}
$, where

$J_2 = \color{red}{V_{ij}V_{ij}}$

There's a complicated expression for $μ$ now (empirical?):

$\color{crimson}{μ(J_2)}
 =\frac{1}{\sqrt{J_2}}
  \left(
    \sqrt[4]{η^2 J_2}
   - \frac{1}{\sqrt{2}} \sqrt{τ_y}
  \right)^2
$

Plugging into conservation of momentum equation:

$\color{magenta}{ ρ\frac{∂v_i}{∂t} }
 = \color{orange}{ -\frac{∂p}{∂x_i}
 + \color{crimson}{μ(J_2)} \color{red}{\frac{∂^2 v_i}{∂x_j^2}}}
 + \color{red}{V_{ij}}\frac{∂}{∂x_j}\color{crimson}{μ(J_2)}
 + χ_i
$

(4 equations in 4 unknowns)

This has extra terms compared to Navier-Stokes.

## Reynold's Number
Back to Navier-Stokes...
* characteristic quantities
* $ N_R
   = \frac{VLρ}{μ}
   = \frac{inertial force}{viscous force}
  $

  * $N_R \gg 1 ⇒$ inertial forces dominate $⇒$ ignore viscosity term (eg. in aorta). Body force $χ_i$ might be 0 too.
   $ \color{magenta}{ ρ\frac{∂v_i}{∂t} }
   = \color{orange}{ -\frac{∂p}{∂x_i} }
   + \color{lightgrey}{μ\frac{∂^2 v_i}{∂x_j^2} }0
   + \color{lightgrey}{χ_i}0 $

   $⇒ \color{magenta}{ ρ\frac{∂v_i}{∂t} }
   + \color{orange}{ \frac{∂p}{∂x_i} }
   = 0$

   missing steps...

   $\begin{equation}\tag{Bernoulli's Equation}
    ⇒ ρ\frac{v_i^2}{2} + p = constant
    \end{equation}
   $

  * $N_R \ll 1 ⇒$ viscous forces dominate $⇒$ ignore momentum (intertial) term  (eg. in capillaries)

   $ \color{lightgrey}{ ρ\frac{∂v_i}{∂t} }
   0 = \color{orange}{ -\frac{∂p}{∂x_i} }
   + μ\color{red}{ \frac{∂^2 v_i}{∂x_j^2} }
   + \color{lightgrey}{χ_i}0
   $

   $\begin{equation}\tag{Stokes Approximation}
    ⇒ \color{orange}{ \frac{∂p}{∂x_i} }
    = μ\color{red}{ \frac{∂^2 v_i}{∂x_j^2} }
    \end{equation}
   $
   and from incompressibility, $\frac{∂{v_i}}{∂x_i}=0$
